```python
lat = pd.read_csv('cbm_train_LATLNG.csv', encoding='cp949', index_col=0)
#lng = pd.read_csv('cbm_test_LATLNG.csv', encoding='cp949', index_col=0)
lat = lat[['pnu', 'Address', 'LAT', 'LNG']]
#lng = lng[['pnu', 'LAT', 'LNG']]
#df = pd.concat([lat, lng],0).reset_index(drop=True)
lat["loc"] = list(zip(lat["LAT"], lat["LNG"]))
lat = lat.replace(r'[<()]', '', regex=True)
lat
```

```python
pg = pd.read_csv('pg.csv', encoding='cp949')
df = pd.merge(pg, lat, how='left', on ='Address')
df
```

```python
from folium.plugins import MarkerCluster

m = folium.Map(
    location = [36.5053542, 127.7043419],
    zoom_start = 8

)
marker_cluster = MarkerCluster().add_to(m)

for i in range(len(df)):
    folium.Marker(
        location = df['loc'][i],
        popup= df['name'][i],
        icon = folium.Icon(color='red', icon='ok')
    ).add_to(marker_cluster)
    
m
```
